```python
#REMOVE HTML TAGS
```

```python
# Import Regular Expression
import re

# Function to remove HTML Tags
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
# Suppose we have a text which contains HTML Tags 
text = "<html><body><p> Movie 1</p><p> Actor - Aamir Khan</p><p> Click here to <a href='http://google.com'>download</a></p></body></html>"
text
```

```python
# Apply the above function to Remove HTML Tags
remove_html_tags(text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
df
```

```python
#Remove URLS
```

```python
# Here we also yse Regular Expressions to remove URLs from the text or whole Corpus.
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
#Removing Punctuations
```

```python
import string
punc = string.punctuation
punc
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('', '', punc))
```

```python
def chat_conversion(text):
    new_text = []
    for i in text.split():
        if i.upper() in chat_words:
            new_text.append(chat_words[i.upper()])
        else:
            new_text.append(i)
    return " ".join(new_text)
```

```python
## Handling Stopwords
##Stop words are common words like "the," "is," and "and," which appear frequently in text but carry little semantic meaning. 
```

```python
from nltk.corpus import stopwords
stopword = stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in stopword:
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
# Text
text = 'probably my all-time favorite movie, a story of selflessness, sacrifice and dedication to a noble cause, but it\'s not preachy or boring. it just never gets old, despite my having seen it some 15 or more times'
print(f'Text With Stop Words :{text}')
# Calling Function
remove_stopwords(text)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
df
```

```python
##Handling Emojis
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
# Texts 
text = "Loved the movie. It was ðŸ˜˜"
text1 = 'Python is ðŸ”¥'
print(text ,'\n', text1)

# Remove Emojies using Fucntion
print(remove_emoji(text))
remove_emoji(text1)
```

```python
##Stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
# Intilize Stemmer
stemmer = PorterStemmer()

# This Function Will Stem Words
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
# A single Sentence
st = "walk walks walking walked"
# Calling the function
stem_words(st)
```

```python
# We Will Import WordNetLemmatizer from NLTK Library.
from nltk.stem import WordNetLemmatizer
# Intilize Lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Sentence 
sentence = "He was running and eating at the same time. He has a bad habit of swimming after playing long hours in the Sun."
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download required resources
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('omw-1.4')

# Initialize
punctuations = "?:!.,;"
wordnet_lemmatizer = WordNetLemmatizer()

# Example sentence
sentence = "He was running and eating at the same time. He has a bad habit of swimming after playing long hours in the Sun."

# Tokenize
sentence_words = nltk.word_tokenize(sentence)

# Remove punctuation safely
sentence_words = [w for w in sentence_words if w not in punctuations]

# Print words and lemma
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v')))
```
