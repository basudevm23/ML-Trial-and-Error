```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"Will is pretty cute")
```

```python
doc.text
```

```python
doc[1].pos_
```

```python
spacy.explain('PROPN')
```

```python
for word in doc:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp(u"Catch me if you can")
for word in doc2:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```
